```python
import datasets
```

```python
data = datasets.load_dataset("xanderios/linkedin-job-postings")
```

```python
data = data.filter(lambda e: "developer" in e['title'].lower(), num_proc=12)
```

```python
data = data.remove_columns(['job_id', 'company_id', 'title', 'max_salary', 'med_salary', 'min_salary', 'pay_period', 'formatted_work_type', 'location', 'applies', 'original_listed_time', 'remote_allowed', 'views', 'job_posting_url', 'application_url', 'application_type', 'expiry', 'closed_time', 'formatted_experience_level', 'skills_desc', 'listed_time', 'posting_domain', 'sponsored', 'work_type', 'currency', 'compensation_type', 'scraped'])
```

```python
data = data.rename_column('description', 'content')
```

```python
topcv = datasets.load_dataset("pre-view/JD-topcv")
```

```python
topcv = topcv.remove_columns(['link'])
```

```python
def filterITJob(example):
    keywords = ["developer", "front", "back", "intern", "senior", "junior", "CNTT"]
    excludes = ["artist", "manager", "analyst", "quản trị"]
    for keyword in excludes:
        if keyword in example['content'].lower():
            return False
    for keyword in keywords:
        if keyword in example['content'].lower():
            return True
    return False

tmp = topcv.filter(filterITJob, num_proc=12)
data = data.filter(filterITJob, num_proc=12)
print(len(tmp['train']))
print(len(data['train']))
```

```python
final_mix = datasets.DatasetDict()
final_mix['train'] = datasets.concatenate_datasets([data['train'], tmp['train']])
```
