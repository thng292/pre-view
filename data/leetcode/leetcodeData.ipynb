```python
import pandas as pd

df = pd.read_parquet('leetcode_problems.parquet')

df = df.dropna()
df = df[df['tags'] != '']
df = df.drop(columns='paid_only')
```

```python
tag = set()
tags = df['tags']
for t in tags:
    for s in t.split(','):
        tag.add(s.strip())

print(len(tag))
tag
```

```python
def MergeTag(row):
    row['tags'] = row['tags'].replace('Binary Indexed Tree', 'Tree').replace('Binary Search Tree', 'Tree').replace('Binary Tree', 'Tree')
    row['tags'] = row['tags'].replace('Topological Sort', 'Sorting').replace('Radix Sort', 'Sorting').replace('Merge Sort', 'Sorting').replace('Counting Sort', 'Sorting').replace('Bucket Sort', 'Sorting')
    row['tags'] = row['tags'].replace('Doubly-Linked List', 'Linked List')
    row['tags'] = row['tags'].replace('String Matching', 'String')
    row['tags'] = row['tags'].replace('Heap (Priority Queue)', "Heap")
    return row

df = df.apply(MergeTag, axis=1)
df
```

```python
tag = set()
tags = df['tags']
for t in tags:
    for s in t.split(','):
        tag.add(s.strip())

print(len(tag))
tag
```

```python
df.to_parquet('leetcode_problems_fix.parquet')
```

```python
# testing

import random
tags = ['Array', 'Hash Table', 'String', 'Design', 'Graph', 'Tree', 'Depth-First Search', 'Breadth-First Search', 'Algorithm']

data = df.to_dict(orient='records')

res = []
weight = []
for d in data:
    w = 0
    for tag in tags:
        if d['tags'].find(tag) != -1: 
            w += 1
    d['match'] = w
    if w > 0:
        res.append(d)
        weight.append(w)

random.choices(res, weights=weight, k=10)
```
