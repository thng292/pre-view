```python
import torch
from datasets import load_from_disk
from IPython.display import Audio

dataset_path = "./data/finalDataset"
dataset = load_from_disk(dataset_path)
sample = dataset[0]

waveform = torch.tensor(sample['audio']['array'])  # Convert the audio array to a torch tensor

sampling_rate = sample['audio']['sampling_rate'] 
print(sample['text'])
Audio(waveform.numpy(), rate=sampling_rate)
```

```python
from collections import Counter

import matplotlib.pyplot as plt

column_name = "duration"
data_column = dataset[column_name]

# Basic statistics
statistics = {
    "count": len(data_column),
    "mean": sum(data_column) / len(data_column) if data_column and isinstance(data_column[0], (int, float)) else None,
    "min": min(data_column),
    "max": max(data_column),
}

# Plot histogram
plt.figure(figsize=(8, 5))
plt.hist(data_column, bins=20, color='blue', alpha=0.7, edgecolor='black')
plt.title(f"Histogram of '{column_name}'")
plt.xlabel(column_name)
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
